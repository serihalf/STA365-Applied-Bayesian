1. Make the "Classic Bayes" figure, but instead of a single likelihood, plot the individual likelihoods of each observation.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# simulate three random samples from a standard normal distribution to setup the likelihood

sample_one = stats.norm(0, 1).rvs(1)
sample_two = stats.norm(0, 1).rvs(1)
sample_three = stats.norm(0, 1).rvs(1)

# setup the likelihood as an array

data = np.array([sample_one, sample_two, sample_three])
likeli_cont = []

# define constants and hyperparameters

n = 3 # number of samples
theta_0, tau = 4, 1/2

# for plotting purpose, generate x data

parameter_domain = np.linspace(-10, 10, 1001)

# store the inidividual likelihood in a container (i.e., generate the likelihood)

for x in data:
    likeli_cont.append(stats.norm.pdf((x[:,np.newaxis])*np.ones(parameter_domain.shape), 
                            loc=parameter_domain, scale=1).prod(axis=0))

# multiply three individual likelihood to make a single likelihood

likelihood = likeli_cont[0]*likeli_cont[1]*likeli_cont[2]

# prior distribution setup

prior = stats.norm(loc=theta_0, scale=1/tau**0.5).pdf(parameter_domain)

# define a constant to scale up the joint 
c = 500

# calculate the joint distribution

joint = c*likelihood*prior

# calculate the posterior distribution

posterior = stats.norm(loc=(tau*theta_0+data.sum())/(tau+n), scale=1/((tau+n)**0.5)).pdf(parameter_domain)

# plot the individual likelihood, prior, posterior and joint distributions

plt.figure(figsize=(16,5))
for i in range(len(likeli_cont)):
    if i == 0:
        plt.plot(parameter_domain, likeli_cont[i], label=f"{i + 1}st likelihood function")
    elif i == 1:
        plt.plot(parameter_domain, likeli_cont[i], label=f"{i + 1}nd likelihood function")
    elif i == 2:
        plt.plot(parameter_domain, likeli_cont[i], label=f"{i + 1}rd likelihood function")
plt.plot(parameter_domain, prior, label="prior")
plt.plot(parameter_domain, posterior, label="posterior")
plt.plot(parameter_domain, joint, linestyle="--", label="joint (times scalar "+str(c)+")")
plt.title("Plots for Likelihood, Joint, Prior, and Posterior distribution of Normal-Normal")
plt.legend()
plt.show()
                       
```
